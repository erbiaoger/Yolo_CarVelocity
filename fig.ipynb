```python
import matplotlib.pyplot as plt
import pathlib
import string

yolo_files = pathlib.Path('/home/lty/MyProjects/yolo/out')
fake_files = pathlib.Path('/home/lty/MyProjects/yolo/fake_B')
raw_files = pathlib.Path('/home/lty/CUT_plus/datasets/car_some_data/testA')

# letter_params = {
#     "fontsize": 12,
#     "verticalalignment": "top",
#     "horizontalalignment": "left",
#     # "bbox": {"edgecolor": "k", "linewidth": 1, "facecolor": "w",}
# }

letter_params = {
    "fontsize": 8,
    "verticalalignment": "top",
    "horizontalalignment": "left",
    "bbox": {"edgecolor": "k", "linewidth": 1, "facecolor": "w",}
}
```

```python

down_list = ['2023-07-22-07-41-58-out.png', 
             '2023-07-22-07-42-58-out.png',
             '2023-07-22-09-05-58-out.png']


up_list = ['2023-07-22-07-45-58-out.png',
           '2023-07-22-07-48-58-out.png',
           '2023-07-22-08-02-58-out.png']

```

```python
from PIL import Image
fig, axes = plt.subplots(3, 3, figsize=(10, 10))

for i in range(3):
    file = raw_files/down_list[i]
    image = Image.open(file)
    new_size = (256, 256)
    resized_image = image.resize(new_size)

    ax = axes[0, i]
    ax.imshow(resized_image, extent=[30, 300, 60, 0], aspect='auto')
    # ax.axis('off')
    ax.text(-0.05, 1.04, string.ascii_lowercase[0*3+i]+')', transform=ax.transAxes, **letter_params)
    if i == 0:
        ax.set_ylabel('Time [s]')
    ax.set_title('Original')
    
    file = fake_files/down_list[i]
    ax = axes[1, i]
    ax.imshow(plt.imread(file), extent=[30, 300, 60, 0], aspect='auto')
    # ax.axis('off')
    ax.text(-0.05, 1.04, string.ascii_lowercase[1*3+i]+')', transform=ax.transAxes, **letter_params)
    if i == 0:
        ax.set_ylabel('Time [s]')
    ax.set_title('Denoise')

    file = yolo_files/down_list[i]
    ax = axes[2, i]
    ax.imshow(plt.imread(file), extent=[30, 300, 60, 0], aspect='auto')
    # ax.axis('off')
    ax.text(-0.05, 1.04, string.ascii_lowercase[2*3+i]+')', transform=ax.transAxes, **letter_params)
    if i == 0:
        ax.set_ylabel('Time [s]')
    ax.set_title('Classification')
    ax.set_xlabel('Distance [m]')

plt.tight_layout()
```
